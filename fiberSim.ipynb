```python
import numpy as np
%pylab inline
```

```python
u = 10               # u subunits per repeat distance
v = 1                # v turns per repeat
c = 1.0              # repeat distance

r0 = (0.5, 0.0, 0.0) # (r, phi, z)
f0 = 1.              # atomic form factor of atom 0
```

```python
# solve l = nv + mu
m = np.arange(-100, 100, 1)
l = np.arange(-20, 20, 1)
n = []
for li in l:
    nl = []
    for mi in m:
        ni = float(li - mi * u) / float(v)
        if ni.is_integer() and np.abs(ni) < u * 3 :# and ni >= 0.0: 
            nl.append(int(ni))
    n.append(nl)            
```

```python
import scipy.special

def G_calc(n, fj, R, zeta, rs_cyl):
    out = np.zeros(R.shape, dtype=np.complex128)
    for i in range(len(fj)):
        out += fj[i] * scipy.special.jv(n, R * rs_cyl[i][0]) \
             * np.exp(-1.0J * n * rs_cyl[i][1]) * np.exp(1.0J * zeta * rs_cyl[i][2])
    return out

def I_calc(ns, ls, Rs, zetas, c, rs_cyl):
    out = np.zeros((zetas.shape[0], Rs.shape[0]), dtype=np.complex128)
    for i, l in enumerate(ls):
        #print '*********************', l, len(ls)
        z = np.argmin(np.abs(zetas - 2. * np.pi * l/ float(c)))
        if z != 0 and z != zetas.shape[0] - 1:
            for ni in ns[i]: 
                for nj in ns[i]:
                    #print ni, nj, zetas[z], z
                    out[z] += G_calc(ni, [1.0], Rs, zetas[z], rs_cyl) * np.conj(G_calc(nj, [1.0], Rs, zetas[z], rs_cyl)) * np.exp(1.0J * (ni - nj) * 0.0)
    return out

# get the cylindrical coords for 3 turns
turns = 3
dphi = 2. * np.pi / (float(u) / float(v))
dz   = float(c) / float(u)
rs_cyl   = []
for i in range(turns * u * v):
    rs_cyl.append([r0[0], i * dphi, i * dz ])
    
rs_car = []
for i in range(len(rs_cyl)):
    x = rs_cyl[i][0] * np.cos(rs_cyl[i][1])
    y = rs_cyl[i][0] * np.sin(rs_cyl[i][1])
    z = rs_cyl[i][2]
    rs_car.append([x,y,z])
    
Rs    = np.arange(-40, 40, 0.1)
#zetas = np.arange(4. * np.pi * l[0] / float(c), 4. * np.pi * l[-1] / float(c), np.pi / (2. * c))
#zetas = np.arange(4. * np.pi * (-7) / float(c), 4. * np.pi * (7) / float(c), np.pi / (2. * c))
zetas = Rs.copy()

I = I_calc(n, l, Rs, zetas, c, rs_cyl)
```
